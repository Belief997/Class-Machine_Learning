```python
import numpy as np
```

```python
def data_load(file_path):
    data_in = np.loadtxt(file_path)
#     print (data_in)
    print(data_in.shape)
    return data_in
```

```python
def data_slice(data_in, type='train'):
    if type == 'train' :
        data_pre=data_in[0:150]  # python 从0开始
    elif type == 'test':
        data_pre=data_in[150:180]
#     print(data_pre)
    print(data_pre.shape)
    return data_pre
```

```python
def data_proc(label, data_pre):
    img_size = 128
    line=0
    for line in  range(0,data_pre.shape[0]) :
        x_center = (data_pre[line][1] + data_pre[line][3])/2
        y_center = (data_pre[line][2] + data_pre[line][4])/2
        length = data_pre[line][3] - data_pre[line][1]
        hight = data_pre[line][4] - data_pre[line][2]
#         print(x_center,y_center,length,hight)
        data_pre[line][0] = label
        data_pre[line][1] = x_center / img_size
        data_pre[line][2] = y_center / img_size
        data_pre[line][3] = length / img_size
        data_pre[line][4] = hight / img_size
    print(data_pre)   
```

```python
def data_save(data , type):
    if type == 'train':
        np.savetxt("train_pre.txt",data, fmt="%f") 
    elif type == 'test':
        np.savetxt("valid_pre.txt",data, fmt="%f")
    print(data.shape)
```

```python

data_in = data_load("bird_gt.txt")
data_pre_0 = data_slice(data_in,type='train')
data_proc( 0, data_pre_0)

data_in = data_load("car_gt.txt")
data_pre_1 = data_slice(data_in,type='train')
data_proc( 1, data_pre_1)

data_in = data_load("dog_gt.txt")
data_pre_2 = data_slice(data_in,type='train')
data_proc( 2, data_pre_2)

data_in = data_load("lizard_gt.txt")
data_pre_3 = data_slice(data_in,type='train')
data_proc( 3, data_pre_3)

data_in = data_load("turtle_gt.txt")
data_pre_4 = data_slice(data_in,type='train')
data_proc( 4, data_pre_4)


data_train=np.vstack((data_pre_0,data_pre_1,data_pre_2,data_pre_3,data_pre_4))
data_save(data_train,'train')
```

```python
data_in = data_load("bird_gt.txt")
data_pre_0 = data_slice(data_in,type='test')
data_proc( 0, data_pre_0)

data_in = data_load("car_gt.txt")
data_pre_1 = data_slice(data_in,type='test')
data_proc( 1, data_pre_1)

data_in = data_load("dog_gt.txt")
data_pre_2 = data_slice(data_in,type='test')
data_proc( 2, data_pre_2)

data_in = data_load("lizard_gt.txt")
data_pre_3 = data_slice(data_in,type='test')
data_proc( 3, data_pre_3)

data_in = data_load("turtle_gt.txt")
data_pre_4 = data_slice(data_in,type='test')
data_proc( 4, data_pre_4)


data_test=np.vstack((data_pre_0,data_pre_1,data_pre_2,data_pre_3,data_pre_4))
data_save(data_test,'test')
```
